### Data Cleaning

````python
#Dropping the duplicate row
df.drop_duplicates(inplace=True)
````

Taking into consideration the following points:

* The column `"homepage"` has most of its values as null.
* The `budget` and `revenue` columns are adjusted as per the `budget_adj` and `revenue_adj` columns.
* The `tagline`, `overview`, `keywords` columns will not be utilized in the analysis.

***Building on the aforementioned, deleting those columns will not impact our anlysis.***

````python
# Drop cloumns of no use
df.drop(['budget', 'revenue', 'homepage', 'tagline', 'overview', 'keywords'], axis=1, inplace=True)
# Verifying successful columns' removal
df.columns
````

````python
#Exploring generes entries
df.genres.value_counts().head(10)
````

As per the previous sample of `genres`' entries, there's a challenge to perform analyses on the individual genre level. In order to do such analyses, the genres have to be separated. To do that, i opted to construct separate dataframe for genres and movie ids to use through the analysis.

````python
# First splitting the multiple genres entries into separate columns.
genres_df = df['genres'].str.split("|", expand=True)

# Verifying successful separation 
genres_df.head(3)
````

````python
# Creating a separate dataframe form unique genres records.
genres_df = genres_df.stack()
genres_df = pd.DataFrame(genres_df)
genres_df.head()
````

````python
#Renaming the genres column and verifying the genres value count
genres_df.rename(columns={0:'genres_adj'}, inplace=True)
genres_df.head()
````

````python
genres_df.genres_adj.value_counts()
````

* [x] ***What are the geners of movies generating the highest revenue?***

````python
# Merging the genres_df with the orignal dataframe to extract the intended insight about individual genres.
df_merged_1 = df.merge(genres_df,left_index=True, right_index=True)
df_merged_1.head()
````

````python
# Droping the original genres column that contains multiple entries
df_merged_1.drop('genres', axis=1, inplace=True)
````

````python
import numpy as np

x = np.array([1,2,3,4])
x
````

````python
x.reshape(4,1)
````

````python
# Which genres generating the highest revenues
df_merged_1.groupby('genres_adj').revenue_adj.sum().sort_values().plot.barh(color='blue', figsize=(10,10), fontsize= 10)
plt.xlabel('Revenue in Dollars', fontsize= 15)
plt.ylabel('Adjusted Genre', fontsize= 15)
plt.title('Total revenues per genre', fontsize=20);
````

* [x] ***What are the geners of movies consuming the highest budgets?***

````python
df_merged_1.groupby('genres_adj').budget_adj.sum().sort_values().plot.barh(color='orange', figsize=(10,10), fontsize= 10)
plt.xlabel('Budget in Dollars', fontsize= 15)
plt.ylabel('Adjusted Genre', fontsize= 15)
plt.title('Total budget per genre', fontsize=20);
````

***It's clear that Action, Adventure, Drame and comedy are the genres generating most revenues and consuming most budgets.***

* [x] ***What are geners of movies recieving the highest vote score?***

````python
df_merged_1.groupby('genres_adj').vote_average.mean().sort_values().plot.barh(color='yellow', figsize=(10,10), fontsize= 10)
plt.xlabel('Average vote', fontsize= 15)
plt.ylabel('Adjusted Genre', fontsize= 15)
plt.title('Average vote per genre', fontsize=20);
````

***On the vote avarage score, Documentary, music, history and animation are receiving the highest average vote average.***

* [x] ***How have the general taste and preferences changed over years?***

````python
#creating a decade column 
bin_edges = [1960,1970,1980,1990,2000,2010,2020]
bin_names = ['60s','70s','80s','90s','00s','10s']
df_merged_1['decade']=pd.cut(df_merged_1['release_year'], bin_edges, labels=bin_names)
df_merged_1.head(2)
````

````python
# Grouping the dataframe by decade and genres and calculating the sum of popularity for each genre across decades
df_decade = df_merged_1.groupby(['decade','genres_adj']).popularity.sum()
# Checking for grouping
df_decade.head(25)
````

````python
# Plotting the progress of popularity for each genre across decades

axes = df_decade.unstack(level=1).plot(kind='line', subplots=True, sharex = False, sharey="all", layout=(5,4), \
                                       figsize=(20,30),xticks=[0,1,2,3,4,5])
fig = plt.gcf()
fig.suptitle("The Popularity of each Genre across Decades", fontsize=25);
````

***The progress of popularity for Drama, Action, Adventure and comedy have been rising dramatically across decade, the matter that can be elaborated by plotting a collective line graph as follows.***

````python
# Plotting the progress of popularity for all genres across decades
df_decade.unstack(level=1).plot(kind='line', subplots=False , figsize=(15,7));
plt.xlabel('Decade', fontsize = 18)
plt.ylabel('Popularity', fontsize = 18)
plt.title("How Different Genres evolved across Decades")
plt.xticks([0,1,2,3,4,5], ['60s','70s','80s','90s','00s','10s']);
````

## Financial insights
As alluded in the introduction, we have almost half of the movies with zero budgets and zero revenues that might be a problem with the data collection or entry. So in counting for the financial figures, it is prudent to limit our analysis to the entries with positive budgets or revenues.

````python
# Detrmining the non_zero movies for budgets and revenues 
non_zero_movies = pd.DataFrame(df.query('revenue_adj > 0 and budget_adj > 0'))
non_zero_movies['profit_loss'] = non_zero_movies['revenue_adj'] - non_zero_movies['budget_adj']

# Total count of non_zero_budget movies
non_zero_movies.shape[0]
````

````python
# Count of movies realising loss
non_zero_movies.query('profit_loss < 0')['original_title'].count()
````

````python
# % of losing movies
non_zero_movies.query('profit_loss < 0')['original_title'].count() / non_zero_movies.shape[0]
````

***So there is 3854 movies with positive budgets and revenues, 1072 of them recording loss which accounts for nearly 28% of the total movies.***

Given that major films costing over 100 million to produce can still flop, this question is more important than ever to the industry.

````python
high_budget =pd.DataFrame(df.query('budget_adj >= 100000000 and revenue_adj > 0'))
high_budget['profit_loss'] = high_budget['revenue_adj'] - high_budget['budget_adj']
high_budget.shape
````

````python
# % of High budget from non_zero_budget movies
high_budget.shape[0] / non_zero_movies.shape[0]
````

````python
# Total count of films of budget equal to or more than 100 million.
total = high_budget.shape[0]
total
````

````python
# Count of losing movies from the high budget films 
high_budget.query('profit_loss < 0')['original_title'].count()
````

````python
# % of losing movies with High budget.
high_budget.query('profit_loss < 0')['original_title'].count() / total
````

***From the above statistics, around 11% of the non zero budget and revenue movies are of budget over 100 million dollars, and 14% of those high budget movies recorded losses.***
